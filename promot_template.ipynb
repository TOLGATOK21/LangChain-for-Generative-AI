```python
from langchain.prompts import PromptTemplate

prompt = PromptTemplate.from_template(
    "Tell me a joke about {topic}"
)
```

```python
from langchain_community.llms import HuggingFaceHub
import os
from langchain.chains import LLMChain

hf = HuggingFaceHub(repo_id = "google/flan-t5-large",
                    model_kwargs={"temperature":0.7,
                                  "max_length":100},
                    huggingfacehub_api_token=os.environ["HUGGINGFACEHUB_API_TOKEN"]
                    )
```

```python
chain = prompt | hf
```

```python
print((chain.invoke({"topic":"You"})))
```

```python
from langchain.prompts import PromptTemplate

template = "You are helpful assistant that translates\{input_language} to {output_language}: {question}"

human_template = "Translate this sentence"
```

```python
prompt = PromptTemplate(
   input_variables=["input_language","output_language","question"],
   template=template
)
```

```python
chain = LLMChain(prompt=prompt, llm=hf)
```

```python
response = chain.run({
    "input_language": "English",
    "output_language": "Turkish",
    "question": "Merhaba benim adım "
})

print(response)
```
